## 1. Estrutura do Sistema de Recomendao

Vamos criar um sistema de recomendao hbrido que combine:

*   **Filtragem Baseada em Contedo:** Recomenda receitas com base nos ingredientes fornecidos pelo usurio, suas preferncias (estilo de vida) e nvel de dificuldade.
*   **Filtragem Colaborativa (em segundo plano):** Aps o projeto estar rodando, utilizar as interaes dos usurios para melhorar as recomendaes ao longo do tempo. Exemplo: um usurio que gosta de um tipo especfico de comida, ser recomendado com mais frequncia.

## 2. Inputs do Usurio

*   **Ingredientes e Data de Validade:** O usurio ir inserir uma lista de ingredientes e suas respectivas datas de validade.
*   **Estilo:** O usurio escolher entre "Normal", "Saudvel" ou "Vegetariano".
*   **Experincia de Cozinha:** O usurio escolher entre "Iniciante", "Intermedirio" ou "Avanado".

## 3. Tratamento dos Dados

### Ingredientes

*   **Lista de Ingredientes Disponveis:** Mapear os ingredientes disponveis no dataset e realizar algum tipo de traduo quando o usurio colocar um ingrediente que no est no dataset.
*   **Data de Validade:** Criar uma lgica para priorizar o uso dos ingredientes com as datas de validade mais prximas.

### Estilo

*   **Normal:** No possui nenhum filtro; poder ser recomendada qualquer receita.
*   **Saudvel:** Filtrar receitas com base nos seus valores nutricionais (priorizar baixa gordura, baixo sdio, etc.). Podemos utilizar as colunas j presentes no dataset.
*   **Vegetariano:** Filtrar apenas receitas classificadas como "Veg" ou "Veg dessert" no dataset.

### Experincia de Cozinha

*   **Iniciante:** Recomendar receitas com menos passos (usando a coluna `n_steps`).
*   **Intermedirio:** Recomendar receitas com um nmero moderado de passos.
*   **Avanado:** Recomendar receitas com um grande nmero de passos. Podemos utilizar a coluna criada anteriormente `difficulty_level`.

## 4. Lgica do Recomendador

*   **Filtragem Inicial:** Filtrar receitas com base no estilo (Normal, Saudvel, Vegetariano) e na experincia de cozinha (Iniciante, Intermedirio, Avanado).
*   **Correspondncia de Ingredientes:**
    *   Verificar quais receitas contm os ingredientes fornecidos pelo usurio.
    *   Calcular uma pontuao de "correspondncia" para cada receita, com base em quantos ingredientes disponveis so utilizados na receita. Por exemplo: Se o usurio possui 4 ingredientes e a receita possui 3 deles, ter uma porcentagem de 75%.
*   **Priorizao por Data de Validade:** As receitas que usam ingredientes com datas de validade mais prximas devem receber uma pontuao mais alta.
*   **Ordenao:** Ordenar as receitas com base na pontuao total (correspondncia de ingredientes + priorizao por data de validade).
*   **Recomendaes:** Retornar as N (por exemplo, 5 ou 10) receitas mais bem pontuadas.

# Sem tradutor

```python
from datetime import datetime
from dateutil import parser
import pandas as pd
import re
import difflib
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer('portuguese')

def classify_difficulty(n_steps):
    if n_steps <= 5:
        return 'Iniciante'
    elif 6 <= n_steps <= 9:
        return 'Intermedirio'
    else:
        return 'Avanado'

def get_closest_match(word, choices):
    matches = difflib.get_close_matches(word, choices, n=1, cutoff=0.7)
    return matches[0] if matches else word

def normalize_word(word):
    return stemmer.stem(word.lower().strip())

def recommend_recipes(df):
    """
    Recomenda receitas com base nos ingredientes disponveis, datas de validade, estilo de vida, experincia e alergias.
    """
    available_ingredients = []
    dates = {}
    allergies = []

    print("Por favor, insira os ingredientes disponveis e suas datas de validade.")
    print("Quando terminar, digite 'fim'.")

    all_ingredients = set(
        ing.strip() for ingredients in df['ingredients'].dropna().astype(str).str.lower().str.replace(r'[^a-zA-Z-\s-]', '', regex=True)
        for ing in ingredients.split(',')
    )

    while True:
        ingredient = input("Ingrediente: ").strip().lower()
        if ingredient == 'fim':
            break
        if not re.match(r'^[a-zA-Z-\s-]+$', ingredient):
            print("Ingrediente invlido. Use apenas letras e espaos.")
            continue

        ingredient = get_closest_match(normalize_word(ingredient), [normalize_word(ing) for ing in all_ingredients])

        while True:
            date_str = input(f"Data de validade para {ingredient} (DD-MM-YYYY): ").strip()
            try:
                parsed_date = parser.parse(date_str, dayfirst=True)
                dates[ingredient] = parsed_date.strftime('%Y-%m-%d')
                break
            except ValueError:
                print("Formato de data invlido. Tente novamente.")

        available_ingredients.append(ingredient)

    print("\nAgora, vamos falar sobre alergias.")
    print("Digite 'fim' para encerrar a lista de alergias.")

    while True:
        allergy = input("Alergia: ").strip().lower()
        if allergy == 'fim':
            break
        if not re.match(r'^[a-zA-Z-\s-]+$', allergy):
            print("Alergia invlida. Use apenas letras e espaos.")
            continue
        allergies.append(allergy)

    valid_lifestyles = {'normal', 'saudvel', 'vegetariano'}
    while True:
        lifestyle = input("\nDigite seu estilo de vida (Normal, Saudvel, Vegetariano): ").strip().lower()
        if lifestyle in valid_lifestyles:
            break
        print("Opo invlida. Escolha entre Normal, Saudvel ou Vegetariano.")

    valid_experience = {'iniciante', 'intermedirio', 'avanado'}
    while True:
        experience = input("Digite sua experincia de cozinha (Iniciante, Intermedirio, Avanado): ").strip().lower()
        if experience in valid_experience:
            break
        print("Opo invlida. Escolha entre Iniciante, Intermedirio ou Avanado.")

    df_filtered = df.copy()

    if 'food types' in df.columns:
        if lifestyle == 'saudvel':
            df_filtered = df_filtered[df_filtered['food types'].str.contains('Healthy', na=False)]
        elif lifestyle == 'vegetariano':
            df_filtered = df_filtered[df_filtered['food types'].str.contains('Veg|Veg dessert', na=False, regex=True)]

    if 'difficulty_level' in df.columns:
        df_filtered = df_filtered[df_filtered['difficulty_level'].str.lower() == experience]

    if allergies and 'ingredients' in df.columns:
        allergy_pattern = '|'.join(map(re.escape, allergies))
        df_filtered = df_filtered[~df_filtered['ingredients'].str.lower().str.contains(allergy_pattern, na=False)]

    if 'ingredients' in df_filtered.columns:
        df_filtered['ingredients'] = df_filtered['ingredients'].astype(str)
        df_filtered['ingredient_match'] = df_filtered['ingredients'].apply(
            lambda x: sum(1 for ing in available_ingredients if ing in x.lower()) / (len(x.split(',')) + 1) if available_ingredients else 0
        )

    max_extra_ingredients = 3  # Limita o nmero de ingredientes extras na receita
    df_filtered = df_filtered[df_filtered['ingredients'].apply(lambda x: len(x.split(',')) - len(available_ingredients) <= max_extra_ingredients)]

    df_filtered['total_score'] = df_filtered['ingredient_match'] * 0.7 + df_filtered['expiration_priority'] * 0.3

    recommended_recipes = df_filtered.sort_values('total_score', ascending=False)
    return recommended_recipes.head(10)


```

```python
from datetime import datetime
from dateutil import parser
import pandas as pd
import re
import difflib
from nltk.stem import SnowballStemmer
from deep_translator import GoogleTranslator
import ast

stemmer = SnowballStemmer('portuguese')
translator_en_to_pt = GoogleTranslator(source='en', target='pt')
translator_pt_to_en = GoogleTranslator(source='pt', target='en')

def translate_list_to_portuguese(value):
    """
    Traduz uma lista de palavras do ingls para o portugus.
    Se `value` for uma string representando uma lista, converte antes de traduzir.
    Se for uma lista real, traduz cada elemento.
    Se for um nico valor, traduz normalmente.
    """
    if isinstance(value, str):
        try:
            # Tenta converter strings que representam listas para listas reais
            parsed_value = ast.literal_eval(value)
            if isinstance(parsed_value, list):
                value = parsed_value
        except (ValueError, SyntaxError):
            # Se no puder ser convertido, trata como string normal
            return translate_to_portuguese(value)

    if isinstance(value, list):
        try:
            return [translate_to_portuguese(item) for item in value]
        except Exception as e:
            print(f"Erro na traduo da lista {value}: {e}")
            return value  # Retorna a lista original em caso de erro

    return value  # Se no for lista nem string, retorna como est



def classify_difficulty(n_steps):
    if n_steps <= 5:
        return 'Iniciante'
    elif 6 <= n_steps <= 9:
        return 'Intermedirio'
    else:
        return 'Avanado'

def get_closest_match(word, choices):
    matches = difflib.get_close_matches(word, choices, n=1, cutoff=0.7)
    return matches[0] if matches else word

def normalize_word(word):
    return stemmer.stem(word.lower().strip())

def translate_to_english(word):
    try:
        return translator_pt_to_en.translate(word)
    except Exception as e:
        print(f"Erro na traduo de '{word}': {e}")
        return word  # Retorna o prprio texto em caso de erro

def translate_to_portuguese(word):
    try:
        return translator_en_to_pt.translate(word)
    except Exception as e:
        print(f"Erro na traduo de '{word}': {e}")
        return word  # Retorna o prprio texto em caso de erro

def recommend_recipes(df, max_recipes=10):
    """
    Recomenda receitas com base nos ingredientes disponveis, datas de validade, estilo de vida, experincia e alergias.
    max_recipes: Nmero mximo de receitas a serem retornadas (padro: 10)
    """
    available_ingredients = []
    dates = {}
    allergies = []

    valid_lifestyles = {'normal', 'saudvel', 'vegetariano'}
    while True:
        lifestyle = input("\nDigite seu estilo de vida (Normal, Saudvel, Vegetariano): ").strip().lower()
        if lifestyle in valid_lifestyles:
            break
        print("Opo invlida. Escolha entre Normal, Saudvel ou Vegetariano.")

    valid_experience = {'iniciante', 'intermedirio', 'avanado'}
    while True:
        experience = input("Digite sua experincia de cozinha (Iniciante, Intermedirio, Avanado): ").strip().lower()
        if experience in valid_experience:
            break
        print("Opo invlida. Escolha entre Iniciante, Intermedirio ou Avanado.")

    print("\nPor favor, insira os ingredientes disponveis e suas datas de validade.")
    print("Quando terminar, digite 'fim'.")

    all_ingredients = set(
        ing.strip() for ingredients in df['ingredients'].dropna().astype(str).str.lower().str.replace(r'[^a-zA-Z-\s-]', '', regex=True)
        for ing in ingredients.split(',')
    )

    normalized_ingredients = {normalize_word(ing): ing for ing in all_ingredients}

    while True:
        ingredient = input("Ingrediente: ").strip().lower()
        if ingredient == 'fim':
            break
        if not re.match(r'^[a-zA-Z-\s-]+$', ingredient):
            print("Ingrediente invlido. Use apenas letras e espaos.")
            continue

        translated_ingredient = translate_to_english(ingredient)  # Traduz para ingls
        normalized_input = normalize_word(translated_ingredient)
        ingredient = get_closest_match(normalized_input, list(normalized_ingredients.keys()))
        ingredient = normalized_ingredients.get(ingredient, ingredient)  # Recupera a verso original

        while True:
            date_str = input(f"Data de validade (Ex: 20-02-2025): ").strip()
            try:
                parsed_date = parser.parse(date_str, dayfirst=True)
                dates[ingredient] = parsed_date.strftime('%Y-%m-%d')
                break
            except ValueError:
                print("Formato de data invlido. Tente novamente.")

        available_ingredients.append(ingredient)

    print("\nAgora, vamos falar sobre alergias. Algumas alergias comuns incluem: leite, ovo, amendoim, frutos do mar, soja, trigo. Se voc tiver outras alergias, digite-as tambm.")
    print("Digite 'fim' para encerrar a lista de alergias.")

    while True:
        allergy = input("Alergia: ").strip().lower()
        if allergy == 'fim':
            break
        if not re.match(r'^[a-zA-Z-\s-]+$', allergy):
            print("Alergia invlida. Use apenas letras e espaos.")
            continue
        allergies.append(translate_to_english(allergy))  # Traduz alergias para ingls

    df_filtered = df.copy()

    if 'food types' in df.columns:
        if lifestyle == 'saudvel':
            df_filtered = df_filtered[df_filtered['food types'].str.contains('Healthy', na=False)]
        elif lifestyle == 'vegetariano':
            df_filtered = df_filtered[df_filtered['food types'].str.contains('Veg|Veg dessert', na=False, regex=True)]

    if 'difficulty_level' in df.columns:
        df_filtered = df_filtered[df_filtered['difficulty_level'].str.lower() == experience]

    if allergies and 'ingredients' in df.columns:
        allergy_pattern = '|'.join(map(re.escape, allergies))
        df_filtered = df_filtered[df_filtered['ingredients'].notna()]
        df_filtered = df_filtered[~df_filtered['ingredients'].str.lower().str.contains(allergy_pattern, na=False)]

    if 'ingredients' in df_filtered.columns:
        df_filtered['ingredients'] = df_filtered['ingredients'].astype(str)
        df_filtered['ingredient_match'] = df_filtered['ingredients'].apply(
            lambda x: sum(1 for ing in available_ingredients if ing in x.lower()) / (len(x.split(',')) + 1) if available_ingredients else 0
        )

    max_extra_ingredients = 3
    df_filtered = df_filtered[df_filtered['ingredients'].apply(lambda x: abs(len(x.split(',')) - len(available_ingredients)) <= max_extra_ingredients)]

    if 'expiration_priority' not in df_filtered.columns:
        df_filtered['expiration_priority'] = 1  # Define um valor padro

    df_filtered['total_score'] = df_filtered['ingredient_match'] * 0.7 + df_filtered['expiration_priority'] * 0.3

    recommended_recipes = df_filtered.sort_values('total_score', ascending=False).head(max_recipes)

    # Traduzindo colunas que podem conter listas ou strings
    recommended_recipes['name'] = recommended_recipes['name'].apply(translate_to_portuguese)
    recommended_recipes['ingredients'] = recommended_recipes['ingredients'].apply(translate_list_to_portuguese)
    recommended_recipes['food types'] = recommended_recipes['food types'].apply(translate_to_portuguese)
    recommended_recipes['steps'] = recommended_recipes['steps'].apply(translate_list_to_portuguese)
    recommended_recipes['description'] = recommended_recipes['description'].apply(translate_to_portuguese)

    return recommended_recipes



```
